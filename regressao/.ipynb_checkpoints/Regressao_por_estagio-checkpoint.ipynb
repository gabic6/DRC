```python
df_estagio1 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][0][1]
df_estagio2 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][1][1]
df_estagio31 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][2][1]
df_estagio32 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][3][1]
df_estagio4 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][4][1]
df_estagio5 = [i for i in df_por_data.groupby('ESTAGIOF_EQ')][5][1]
```

```python
array_df = [df_estagio1, df_estagio2, df_estagio31, df_estagio32, df_estagio4, df_estagio5]
```

## Normalização

```python
array_df_norm = []
for df_est in array_df:
    scaler = preprocessing.MinMaxScaler( feature_range=(-1, 1) )
    reg_mul_norm = scaler.fit_transform(df_est)
    reg_mul_norm = pd.DataFrame(reg_mul_norm,columns=df_est.keys())
    array_df_norm.append(reg_mul_norm)
array_df_norm[4]
# array_df_norm = array_df
```

### Separando em conjunto de teste e treinamento

```python
from sklearn.model_selection import train_test_split
xtrain_df = []
ytrain_df = []
xtest_df = []
ytest_df = []
for est in array_df_norm:
    X = est.drop(columns='ESTAGIOF_EQ')
    y = est['ESTAGIOF_EQ']
    xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.30,random_state= 20200301)
    
    xtrain_df.append(xtrain)
    ytrain_df.append(ytrain)
    xtest_df.append(xtest)
    ytest_df.append(ytest)
```
